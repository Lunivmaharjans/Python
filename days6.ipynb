```python
class Student:
    def __init__(self,name, rollno, marks):
        self.name = name 
        self.rollno = rollno
        self.marks = marks

    def get_average(self):
        return sum(self.marks) / len(self.marks)
    
    def add_mark(self,mark):
        return self.marks.append(mark)
    
    def is_pass(self):
        if self.get_average() >= 40:
            return "Pass"
        else:
            return "Fail"
        
    def is_passmark(self):
        if self.get_average()>=50 and all( m >= 40 for m in self.marks):
            return "Pass"
        else:
            return "Fail"
        
    
        
        


```
