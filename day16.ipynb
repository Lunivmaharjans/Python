```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 500)

# Calculate y values
y = 1 / (1 + np.exp(-x+1))

plt.plot(x, y)

# Labels and title
plt.title("Graph of y = 1 / (1 + e^-x+1)")
plt.xlabel("x")
plt.ylabel("y")

# Grid
plt.grid(True)

# Show plot
plt.show()
```
