```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def predict_y(X, w, b):
    m = X.shape[0]
    y_probs = np.zeros(m)

    for i in range(m):
        y_probs[i] = sigmoid(w * X[i] + b)

    # FIX: comma missing in np.where
    y_preds = np.where(y_probs >= 0.5, 1, 0)

    return y_probs, y_preds
```
