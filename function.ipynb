```python
def fibonacci(n):
    result = []
    a , b = 1,1
    result.append(a)
    result.append(b)
    for _ in range(1,n-1): 
        c = a + b
        a , b = b , c 
        result.append(c) 
    return result
fibonacci(6)
```
