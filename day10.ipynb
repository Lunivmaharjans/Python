```python
a = int(input("enter a number"))

if a % 2 == 0:
    print("it is even")
else:
    print("it is odd")
```

```python
if a <= 1:
    print("The number is not Prime")
else:
    for i in range(2, a):
        if a % i == 0:
            print("The number is not Prime")
            break
    else:
        print("The number is Prime")
```

```python
a = abs(a)

sum_digits = 0


while a > 0:
    digit = a % 10
    sum_digits += digit
    a = a // 10

print("Sum of digits =", sum_digits)
```
